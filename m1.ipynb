```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/aaryamanshinde/Desktop/python/cursor_project/Jobs_NYC_Postings_20241004.csv')
```

```python
df = pd.DataFrame(df)
```

```python
# Recommendation function based on user input
def recommend_jobs(input_skills):
    input_skills = input_skills.lower().split(',')  # Convert input to lowercase and split by commas
    input_skills = [skill.strip() for skill in input_skills]  # Strip whitespace from each skill

    # Initialize an empty list to store job recommendations
    recommendations = []

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        job_description = str(row['Job Description']).lower()  # Convert job description to lowercase
        
        # Check if any of the input skills are in the job description
        if any(skill in job_description for skill in input_skills):
            recommendations.append(row)

    # Display the recommendations
    if recommendations:
        recommended_df = pd.DataFrame(recommendations)
        print("\nRecommended Jobs based on your skills:")
        print(recommended_df[['Job ID', 'Business Title', 'Job Description']])  # You can add more columns if needed
    else:
        print("\nNo matching jobs found. Please try with different skills.")

```

```python
# Example usage - ask for user input
input_skills = input("Enter the skills or keywords (comma-separated): ")

# Call the recommendation function
recommend_jobs(input_skills)
```

```python
import spacy
# Load pre-trained SpaCy model for English
nlp = spacy.load('en_core_web_sm')
```

```python
# Define a function to extract skills from a job description
def extract_skills_nlp(job_description):
    # Check if the job description is a string
    if isinstance(job_description, str):
        # Process the job description with the SpaCy NLP model
        doc = nlp(job_description)
        
        # Example skills list for demonstration (you can expand this)
        skills = [
            "AutoCAD", "Python", "Java", "C++", "Machine Learning",
            "Data Analysis", "Project Management", "Communication",
            "Problem Solving", "Engineering", "Statistics", "SQL",
            "Cybersecurity", "Networking", "CAD", "3D Modeling"
        ]
        
        # Extract skills found in the job description
        extracted_skills = [skill for skill in skills if skill.lower() in job_description.lower()]
        return extracted_skills
    else:
        # Return an empty list if the job description is not a string
        return []

# Assuming df is your existing DataFrame with job postings
# Apply the function to your DataFrame
df['extracted_skills'] = df['Job Description'].apply(extract_skills_nlp)

# Display the updated DataFrame
print(df[['Job ID', 'Business Title', 'extracted_skills']])
```

```python
def recommend_jobs(input_skills):
    # Convert input skills to a list and strip whitespace, then normalize to lowercase
    input_skills = [skill.strip().lower() for skill in input_skills.split(',')]
    
    # Filter DataFrame for jobs that contain any of the input skills (also normalized to lowercase)
    recommended_jobs = df[df['extracted_skills'].apply(
        lambda x: any(skill.lower() in (s.lower() for s in x) for skill in input_skills)
    )]
    
    return recommended_jobs[['Job ID', 'Business Title', 'extracted_skills']]

# Example usage
input_skills = input("Enter skills (comma-separated): ")
recommended = recommend_jobs(input_skills)

# Display recommendations
print(recommended)
```
